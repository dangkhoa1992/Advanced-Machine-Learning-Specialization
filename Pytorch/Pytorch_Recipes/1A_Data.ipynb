# 1. Data

```python
import torch
import torchaudio
import matplotlib.pyplot as plt
```

#### Download data

```python
yesno_data = torchaudio.datasets.YESNO(
    root="/tmp",
    url='http://www.openslr.org/resources/1/waves_yesno.tar.gz',
    folder_in_archive='waves_yesno',
    download=True)
```

```python
# Examine the data[0]
waveform, sample_rate, labels = yesno_data[0]
print(f'''Wavform shape: {waveform.shape}
Sample_rate: {sample_rate}
Label (1: Yes, 0: No): {labels}''')

plt.figure()
plt.plot(waveform.t().numpy());
```

#### Feed data to dataloader

```python
data_loader = torch.utils.data.DataLoader(
    yesno_data,
    batch_size=1,
    shuffle=True)
```

```python
# Examine 1st batch of dataloader
for data in data_loader:
    waveform, sample_rate, labels = data[0], data[1], data[2]
    print(f'''Wavform shape: {waveform.shape}
Sample_rate: {sample_rate}
Label (1: Yes, 0: No): {labels}''')

    plt.figure()
    plt.plot(waveform[0].t().numpy())
    break
```
