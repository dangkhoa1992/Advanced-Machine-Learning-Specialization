# 1. Sums of Normals


#### Define population
- $X \sim Normal(\mu, \sigma^2)$

#### Define sample spaces

+ 1 sample consists of: $\{ x_1, x_2, \dots, x_N \}$
    + sameple mean: $\hat{\mu} = \bar{x} = \frac{1}{N} \sum\limits_{i=1}^N x_i $
    + sample variance: $\hat{\sigma}^2 = \frac{1}{N-1} \sum\limits_{i=1}^N (x_i - \bar{x})^2$

- Let take multiple samples from the population
    + Each sample is `IID` (independent, identically distributed)

- Take samples from population
    + $X_1 \sim Normal(\mu_1, \sigma_1^2)$
    + $X_2 \sim Normal(\mu_2, \sigma_2^2)$
    + $\dots$
    - Each $X_i$ is iid to each others

## 1.1 Sums of Normals problem
- Take samples (sample size = N) from the population
- Let $\hat{\mu}$ is the random var represent the mean of all samples (the mean of the estimate)
    + Find its distribution

#### Solve
- The mean of the estimate

$$\begin{split}
E(\hat{\mu}) &= E(\frac{1}{N} \sum\limits_{i=1}^N X_i) \\
    &= E\left( \frac{1}{N} (X_1 + X_2 + \dots + X_N)\right) \\
    &= \frac{1}{N}\mu_1 + \frac{1}{N}\mu_2 + \dots + \frac{1}{N}\mu_N \\
    &= \mu
\end{split}$$

- The variance of the estimate 

$$\begin{split}
Var(\hat{\mu}) &= Var(\frac{1}{N} \sum\limits_{i=1}^N X_i) \\
    &= E \left[ \left( \frac{1}{N} \sum\limits_{i=1}^N (X_i - \mu) \right)^2 \right] \\
    &= \frac{1}{N^2} E \left[ \left( \sum\limits_{i=1}^N (X_i - N \mu) \right)^2 \right] \\
    &= \frac{1}{N^2} Var \left( \sum\limits_{i=1}^N X_i \right) \\
    &= \frac{1}{N^2} N \sigma^2 \\
    &= \frac{\sigma^2}{N}
\end{split}$$

## 1.2 Central Limit Theorem

- Conclusion about the distribution of $\hat{\mu}$ (sample mean): $\hat{\mu} \sim Normal(\mu, \frac{\sigma^2}{N})$
    + The mean of all samples is the population mean
    + The variance of the estimate depend on population var and how many samples we take

- If X ~ other than normal distribution, always $\hat{\mu} \sim Normal(\mu, \frac{\sigma^2}{N})$
    + The mean of samples is `always normal distributed`

# 2. $\gamma$ - confidence interval

- Meaning: 
    + $\gamma$ of the time, the $\gamma$ confidence interval will contain the true $\mu$
    + Or If we do an experiment many times, $\gamma$ \% of those time the population $\mu$ will contained within that CI
    
$$\gamma = \frac{\text{Number of experiments where } \mu \text{ is contained in the } \gamma \text{-CI}}{\text{number of total experiments}}$$
- Common gamma value: 95\%, 90\%, 99\% 

- Example: 95\% Confidence Interval, $\gamma = 0.95$
    - Meaning: 95\% of the time, the 95\% confidence interval will contain the true $\mu$

## 2.1 $\gamma$
- Confidence interval (CI) is defined as
    + the area that the interval cover $\gamma$ percent over the area of an distribution


- Example: 95\% z-CI
    + The interval will cover 95\% middle area (green) of the normal distribution
    + Or CI = P(lower_limit $\leq x \leq$ upper_limit )

## 2.2 Z -  confidence interval
- Apply for `normal distribution`
- z-distribution is `standard normal distribution`

- Z-CI is the area covered by $P(z_{left} \leq z \leq z_{right})$

<img src="assets/3.jpg" width="400"/>


#### Solve Standard normal distribution  $\gamma = 0.95$
- Suppose we working with the standard normal distribution: z ~ N(0,1)
    + $\gamma = 0.95 = P(z_{left} \leq z \leq z_{right})$
    

- Calc red area
    + red area + $\gamma$ + red area = 1 => red area = $\frac{1 - \gamma}{2}$ 

- $p_{low} = \text{ red area } = \frac{1 - \gamma}{2} = 0.025$
    + $z_{left} = F^{-1}(p_{low}) = ppf(p_{low}) = ppf \left(\frac{1 - \gamma}{2} \right) = ppf(0.025) = -1.96$

- $p_{high} = \gamma + \text{ red area } = \frac{1 + \gamma}{2} $
    + $z_{right} = F^{-1}(p_{high}) = ppf(p_{high}) = ppf \left(\frac{1 + \gamma}{2} \right) = ppf(0.975) = 1.96$
    
- 95\%-CI of z = \[ -1.96, 1.96 \]

#### Solve Normal distribution  $\gamma = 0.95$
- Suppose we working with the normal distribution: x ~ $N(\mu, \sigma_0^2)$
    + standarize x to z as: $z = \frac{x - \mu}{\sigma_0} \sim N(0,1)$
    + $\gamma = 0.95$

- We have 95\%-CI of z = \[ -1.96, 1.96 \]

$$\begin{split}
-1.96 &\leq z \leq 1.96 \\
-1.96 &\leq \frac{\hat{\mu} - \mu}{\sigma_0} \leq 1.96 \\
\hat{\mu} - 1.96\sigma_0 &\leq \mu \leq \hat{\mu} + 1.96\sigma_0
\end{split}$$

- From `Sums of Normals` findings:
    + If $X \sim N(\mu, \sigma^2)$ then $\hat{\mu} \sim N(\mu, \frac{\sigma^2}{N})$

- 95\%-CI of z = $\left[ \hat{\mu} -1.96\frac{\sigma}{\sqrt{N}}, \hat{\mu} + 1.96 \frac{\sigma}{\sqrt{N}} \right]$


#### Generic formulas for $\gamma$ 
- Let
    + $\hat{\mu}$: mean of the sample
    + $\sigma$: std of the population
        + We don't know but can be replaced by $\hat{\sigma}$ (sample std)
        + $\hat{\sigma} = \sqrt{ \frac{1}{N-1} \sum\limits_{i=1}^N (x_i - \bar{x})^2 }$
    + $X \sim N(\mu, \sigma^2)$, then $\hat{\mu} \sim N(\mu, \frac{\sigma^2}{N} )$
    + $F^{-1}(p) = \text{norm.ppf(p)}$

$$\gamma-\text{CI} = \left[  \hat{\mu} + \text{norm.ppf}\left( \frac{1-\gamma}{2} \right) \frac{\sigma}{\sqrt{N}}, \hat{\mu} + \text{norm.ppf}\left( \frac{1+\gamma}{2} \right) \frac{\sigma}{\sqrt{N}}  \right]$$

- In practice, replace $\sigma$ by $\hat{\sigma}$
    + $z_{\text{left}} = F^{-1}\left( \frac{1-\gamma}{2} \right) = \text{norm.ppf}\left( \frac{1-\gamma}{2} \right)$
    + $z_{\text{right}} = F^{-1}\left( \frac{1+\gamma}{2} \right) = \text{norm.ppf}\left( \frac{1+\gamma}{2} \right)$


$$\gamma-\text{CI} = \left[  \hat{\mu} + \text{norm.ppf}\left( \frac{1-\gamma}{2} \right) \frac{\hat{\sigma}}{\sqrt{N}}, \hat{\mu} + \text{norm.ppf}\left( \frac{1+\gamma}{2} \right) \frac{\hat{\sigma}}{\sqrt{N}}  \right]$$

#### Notes
- According to `Central Limit Theorem`, The mean of samples is `always normal distributed` --> we can apply z-CI for non-normal distributed data

## 2.3 t-confident interval
- Issue with z-CI:
    + We dont know $\sigma$
    + Assume $\sigma = \hat{\sigma}$


#### $\chi^2$-distribution (chi-squared)
- Let $z_1, z_2, \dots, z_i, \dots, z_k$ are `independent`, `standard normal` random variables
- Define Chi-square distribution as
    + `k`: degree of freedom

$$Q = \sum\limits_{i=1}^k z_i^2 \sim \chi^2(k)$$



#### t-distribution

- Let t is a new standardized random variable (t is not normally distributed)

$$t = \frac{\hat{\mu} - \mu}{\hat{\sigma} / \sqrt{N}} $$


- Transform t
$$\begin{split}
t &= \frac{\hat{\mu} - \mu}{\hat{\sigma} / \sqrt{N} } \\
    &= \frac{\hat{\mu} - \mu}{\sigma / \sqrt{N} } * \frac{\sigma}{\hat{\sigma}}
\end{split}$$

- We have
    + $\frac{\hat{\mu} - \mu}{\sigma / \sqrt{N} }$ is normally distributed ($\sim N$)
    + $\frac{\sigma}{\hat{\sigma}}$ is Chi-square distributed with Degree of freedom = N-1 ($\sim \chi^2 (N-1)$)
    
$$\begin{split}
\left ( \frac{\hat{\sigma}}{\sigma} \right)^2 &= \frac{1}{N - 1} \sum\limits_{i=1}^N \left( \frac{x_i - \bar{x}}{\sigma} \right)^2 \\
(N - 1) \left ( \frac{\hat{\sigma}}{\sigma} \right)^2 &= \sum\limits_{i=1}^N \left( \frac{x_i - \bar{x}}{\sigma} \right)^2 \sim \chi^2_{N-1}
\end{split}$$

- We define t-distribution as 
    + $t_{\text{degree of freedom}} \sim \frac{\text{standard normal} }{\sqrt{\text{chi-square / degree of freedom}}} $ 

$$t = \frac{z}{\sqrt{V/\nu}} \sim t_{\nu}$$

<img src="assets/4.jpg" width="400"/>

- Intuition
    + t-distribution = a detailed modification of z-distribution
    + t-distribution has a bell-curve shape like normal distribution (In z-CI, we did assume $\hat{\sigma} = \sigma$)
    + t-distribution has flatter bell curve than z-distribution -> t-CI will be wider than z-CI
    + As N become large, t becomes z

- Formula (Note: df = degree of freedom)
    + $t_{\text{left}} = F^{-1}\left( \frac{1-\gamma}{2} \right)_{df = N-1} = \text{t.ppf}\left( \frac{1-\gamma}{2} \right)_{df = N-1}$
    + $t_{\text{right}} = F^{-1}\left( \frac{1+\gamma}{2} \right)_{df = N-1} = \text{t.ppf}\left( \frac{1+\gamma}{2} \right)_{df = N-1}$


$$\gamma-\text{CI} = \left[  \hat{\mu} + \text{t.ppf}\left( \frac{1-\gamma}{2} \right)_{df = N-1} \frac{\hat{\sigma}}{\sqrt{N}}, \hat{\mu} + \text{t.ppf}\left( \frac{1+\gamma}{2} \right)_{df = N-1} \frac{\hat{\sigma}}{\sqrt{N}}  \right]$$
