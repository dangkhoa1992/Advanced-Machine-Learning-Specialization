

# 1. Gausian distribution
- For model like: rating, amount of time user spends on the web page

## 1.1 Example
- Collect height data in a classroom as: $X = \{x_1, x_2, \dots, x_N \}$, $x_i \in R$

- Assume the data is normal distributed which pdf
    + $p(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-1}{2} \left(  \frac{x - \mu}{\sigma} \right)^2}$

- Find the mean, variance of this distribution

#### Solve

- Likelihood function
    
$$\begin{split}
L(\theta)  &= \prod\limits_{i=1}^N p(x_i | \theta),\text{ where } \theta = \{ \mu, \sigma^2 \} \\
    &= \prod\limits_{i=1}^N \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-1}{2} \left(  \frac{x_i - \mu}{\sigma} \right)^2} \\
\end{split}$$

- Log-Likelihood function

$$\begin{split}
log L(\theta)  &= log \prod\limits_{i=1}^N p(x_i | \theta),\text{ where } \theta = \{ \mu, \sigma^2 \} \\
    &= log \prod\limits_{i=1}^N \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-1}{2} \left(  \frac{x_i - \mu}{\sigma} \right)^2} \\
    &= \sum\limits_{i=1}^N log \left(   \frac{1}{\sigma \sqrt{2\pi}}e^{\frac{-1}{2} \left(  \frac{x_i - \mu}{\sigma} \right)^2} \right) \\
    &= \sum\limits_{i=1}^N log \left(   \frac{1}{\sigma \sqrt{2\pi}}  \right) + \sum\limits_{i=1}^N log \left ( e^{{\frac{-1}{2} \left(  \frac{x_i - \mu}{\sigma} \right)^2}} \right) \\
    &=  \frac{-1}{2}log(2\pi\sigma^2) \sum\limits_{i=1}^N 1 + \sum\limits_{i=1}^N  {\frac{-1}{2} \left(  \frac{x_i - \mu}{\sigma} \right)^2} \\
    &= \frac{-N}{2}log(2\pi\sigma^2) + \frac{-1}{2} \sum\limits_{i=1}^N  { \left(  \frac{x_i - \mu}{\sigma} \right)^2}
\end{split}$$

- Solve $\frac{ d log L(\theta) }{d \theta} = 0$

$$
\frac{ d log L(\theta) }{d \theta} = 0 \\
    => \begin{cases}
    \frac{\delta log L(\mu)}{\delta \mu} = 0 \\
    \frac{\delta log L(\sigma)}{\delta \sigma} = 0 \\
\end{cases}$$

- Solve $\frac{\delta log L(\mu)}{\delta \mu} = 0$

$$\begin{split}
\frac{\delta log L(\mu)}{\delta \mu} &= 0 \\
=> \sum\limits_{i=1}^N  \left(  \frac{x_i - \mu}{\sigma} \right) \frac{1}{\sigma} &= 0 \\
=> \hat{\mu} &= \frac{1}{N} \sum\limits_{i=1}^N x_i
\end{split}$$


- Solve $\frac{\delta log L(\sigma)}{\delta \sigma} = 0$

$$\begin{split}
=> \hat{\sigma}^2 &= \frac{1}{N} \sum\limits_{i=1}^N (x_i - \mu)^2
\end{split}$$

## 1.2 Regression to the mean
- Interestingly, 
    + optimal mean = sample mean
    + optimal variance = sample variance

- $\hat{\mu}$ is also distributed based on different sample data collected, and
    + $\hat{\mu} \approx \text{Normal }(\mu, \frac{\sigma^2}{N})$
    + $E(\hat{\mu}) = \mu$

- Notes: Only $E(\hat{\mu}) = \mu$ 
    + $E(\hat{\sigma}^2) \neq \sigma^2$
    + $E(\hat{\sigma}^2) = \frac{N-1}{N} \sigma^2$

#### Regression to the mean
- Expected value of the sample means is the mean of the random variable: $E(\hat{\mu}) = E(X) = \mu$
- Unbiased Estimate is defined as

$$\hat{\sigma}^2= \frac{1}{N-1} \sum\limits_{i=1}^N (x_i - \hat{\mu})^2$$

#### Side notes
- In `Stats`, $\frac{1}{1-N}$ is different from $\frac{1}{N}$
- In `Machine Learning`, as $N \to \infty$, $N \approx N-1$

$$\hat{\sigma}^2= \frac{1}{N} \sum\limits_{i=1}^N (x_i - \hat{\mu})^2$$


# 2. cdf and ppf
- Let
    + pdf(x) = p(x)
    + cdf(x) = F(x)
- Example
    - A population has normal random distribution of height X ~ norm($\mu, \sigma$) with
        + mean = 170 cm
        + std = 7cm 

#### cdf
- cdf(x) = the probability(Or percentile) that X less or equal to x

$$F(x) = P(X \leq x) = \int\limits_{-\infty}^x p(x)dx $$

- Example: you are 160 cm tall, what percentile are you in?
    + `Ans`: $F(x=160)$

- Example: you are 160 cm tall, what is the probability that someone is taller than you?
    + `Ans`: $1 - F(x=160)$

#### Percentile (ppf)
- Percentile = percent point function (ppf)
    + = the value x at probability (or percentile) p
    + = Inverse function of cdf

$$ ppf(p) = F^{-1}(p)$$

- Example: at what height are you in the 95th percentile?
    + `Ans`: $ppf(p=0.95) = F^-1(p=0.95)$
