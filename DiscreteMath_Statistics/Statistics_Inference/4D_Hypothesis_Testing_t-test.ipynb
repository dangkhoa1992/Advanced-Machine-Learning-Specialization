- Let
    + $\hat{\mu}$ = sample mean
    + $\hat{\sigma}$ = sample std
    + $\mu$ = population mean
    + $\sigma$ = population std
    

#### t-statistic vs z-statistic
- z-statistic
    + We dont know $\sigma$
    + For calculation, Assume $\sigma = \hat{\sigma}$
$$z = \frac{\hat{\mu} - \mu_0}{\sigma / \sqrt{N}} \sim N(0, 1)$$

- t-statistic
$$t = \frac{\hat{\mu} - \mu_0}{\hat{\sigma} / \sqrt{N} } \sim t_{N-1}$$
    


# 1. 1-sample t-test

- Calculate t
$$t = \frac{\hat{\mu} - \mu_0}{\hat{\sigma} / \sqrt{N} }$$

- Calculate degree of freedom
$$\nu = N - 1$$

## 1.1 1-sample, 2-sided test
- 1-sample, 2-sided test
    - $H_0$: $\mu = \mu_0$
    - $H_1$: $\mu \neq \mu_0$
- Calculate p-value
    - $p_{\text{right}} = 1 - \text{t.cdf}(|t|)_{df = N-1} $
    - $p_{\text{left}} = \text{t.cdf}(-|t|)_{df = N-1} $
    - $p = p_\text{left} + p_\text{right}$
    

## 1.2 1-sample, 1-sided test

- 1-sample test, 1-sided test
    - $H_0$: $\mu \leq \mu_0$
    - $H_1$: $\mu > \mu_0$
    
- Calculate p-value
    - $p = 1 - \text{t.cdf}(t)_{df = N-1} $

# 2. 2-sample t-test

## 2.0 3 cases to calculate t and $\nu$

#### Case 1: $N_1 = N_2 = N$ and $\sigma_1 = \sigma_2$ 
- Calculate t
$$s_p^2 = \frac{\hat{\sigma}_1^2 + \hat{\sigma}_2^2}{2} $$

$$t = \frac{\hat{\mu}_1 - \hat{\mu}_2}{ s_p \sqrt{\frac{2}{N}} }$$

- Calculate degree of freedom

$$\nu = 2N-2$$

#### Case 2: $N_1 \neq N_2$ and $\sigma_1 = \sigma_2 = \sigma$ 
- Calculate t
$$s_p^2 = \frac{(N_1 - 1)\hat{\sigma}_1^2 + (N_2 - 1)\hat{\sigma}_2^2}{N_1 + N_2 - 2} $$

$$t = \frac{\hat{\mu}_1 - \hat{\mu}_2}{ s_p \sqrt{\frac{1}{N_1} + \frac{1}{N_2}} }$$

- Calculate degree of freedom

$$\nu = N_1 + N_2 -2$$

#### Case 3: $N_1 \neq N_2$ and $\sigma_1 \neq \sigma_2$ 

- Calculate t
$$t = \frac{\hat{\mu}_1 - \hat{\mu}_2}{ \sqrt{\frac{\hat{\sigma}_1^2}{N_1} + \frac{\hat{\sigma}_2^2}{N_2} }}$$

- Calculate degree of freedom

$$\nu = \frac{\left( \frac{\hat{\sigma}_1^2}{N_1} + \frac{\hat{\sigma}_2^2}{N_2} \right)^2}{\frac{(\hat{\sigma}_1^2 / N_1)^2}{N_1 - 1} + \frac{(\hat{\sigma}_2^2 / N_2)^2}{N_2 - 1}} $$


## 2.1 2-sample, 2-sided test
- 2-sample, 2-sided test
    - $H_0$: $\mu_1 = \mu_2$
    - $H_1$: $\mu_1 \neq \mu_2$
    
- Calculate p-value
    - $p_{\text{right}} = 1 - \text{t.cdf}(|t|)_{df = \nu} $
    - $p_{\text{left}} = \text{t.cdf}(-|t|)_{df = \nu} $
    - $p = p_\text{left} + p_\text{right}$   

## 2.2 2-sample, 1-sided test
- 1-sample test, 1-sided test
    - $H_0$: $\mu_1 \leq \mu_2$
    - $H_1$: $\mu_1 > \mu_2$
    
- Calculate p-value
    - $p = 1 - \text{t.cdf}(t)_{df = \nu} $
