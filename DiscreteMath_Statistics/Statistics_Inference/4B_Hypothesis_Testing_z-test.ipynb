#### Recall from Central Limit Theorem
- Let
    + $\hat{\mu}$ = sample mean
    + $\hat{\sigma}$ = sample std
    + $\mu$ = population mean
    + $\sigma$ = population std

- CLT

$$\hat{\mu} \sim N(\mu, \frac{\sigma^2}{N})$$

# 1. 1-sample z-test
## 1.1 1-sample, 2-sided test
- 1-sample, 2-sided test
    - $H_0$: $\mu = \mu_0$
    - $H_1$: $\mu \neq \mu_0$

- From CLT, we can rewrite the hypotheses as
    - $H_0$: $\hat{\mu} \sim N (\mu_0, \frac{\sigma^2}{N})$
    
- Standarize to z
    - $H_0$: $z = \frac{\hat{\mu} - \mu_0}{\sigma / \sqrt{N}} \sim N(0, 1)$
    
- Assume $\sigma = \hat{\sigma}$
    - $H_0$: $z = \frac{\hat{\mu} - \mu_0}{\hat{\sigma} / \sqrt{N}} \sim N(0, 1)$
    
- Transform to Confidence Interval problem
    + If $H_0$ is true then z will fall into the 95\% - area (assume 95\%-CI)
    + `Rejecting Null Hypothesis` = z is `very unlikely` = fall into the left/right 2.5\% tails
    
<img src="assets/13.jpg" width="400"/>

- Calculate p-value
    - $p_\text{right} = 1 - \text{norm.cdf}(|z|)$
    - $p_\text{left} = \text{norm.cdf}(-|z|)$
    - $p = p_\text{left} + p_\text{right}$ (blue areas)

## 1.2 1-sample, 1-sided test
- 1-sample test, 1-sided test
    - $H_0$: $\mu \leq \mu_0$
    - $H_1$: $\mu > \mu_0$

- If `Rejecting Null Hypothesis` -> $\mu > \mu_0$ then $z = \frac{\hat{\mu} - \mu_0}{\sigma / \sqrt{N}}$ will be large and positive 

- Assume $\sigma = \hat{\sigma}$
    - $z = \frac{\hat{\mu} - \mu_0}{\hat{\sigma} / \sqrt{N}} \sim N(0, 1)$
  
- Calculate p-value
    - $p = 1 - \text{norm.cdf}(z)$

# 2. 2-sample z-test
## 2.1 2-sample, 2-sided test
- 2-sample, 2-sided test
    - $H_0$: $\mu_1 = \mu_2$
    - $H_1$: $\mu_1 \neq \mu_2$
    
- Transform to 1-sample test. **Note**: can choose abitrary $\mu_0$, eg.: 0
    + $H_0$: $y = \mu_1 - \mu_2 = \mu_0$
    + $H_1$: $y = \mu_1 - \mu_2 \neq \mu_0$
    
- Standarize to z. **Note**: $\hat{y} = \hat{\mu_1} - \hat{\mu_2}$
    - $H_0$: $z = \frac{\hat{y} - \mu_0}{\sqrt{\frac{\sigma_1^2}{N_1} + \frac{\sigma_2^2}{N_2}}} \sim N(0, 1)$

- Assume $\sigma_1 = \hat{\sigma_1}$, $\sigma_2 = \hat{\sigma_2}$. **Note**: $\hat{y} = \hat{\mu_1} - \hat{\mu_2}$
    - $H_0$: $z = \frac{\hat{y} - \mu_0}{\sqrt{\frac{\hat{\sigma_1}^2}{N_1} + \frac{\hat{\sigma_2}^2}{N_2}}} \sim N(0, 1)$
    
- Calculate p-value
    - $p_\text{right} = 1 - \text{norm.cdf}(|z|)$
    - $p_\text{left} = \text{norm.cdf}(-|z|)$
    - $p = p_\text{left} + p_\text{right}$
    

## 2.2 2-sample, 1-sided test
- 1-sample test, 1-sided test
    - $H_0$: $\mu_1 \leq \mu_2$
    - $H_1$: $\mu_1 > \mu_2$

- Transform to 1-sample test. **Note**: can choose abitrary $\mu_0$, eg.: 0
    + $H_0$: $y = \mu_1 - \mu_2 \leq \mu_0$
    + $H_1$: $y = \mu_1 - \mu_2 > \mu_0$
    
- Standarize to z
    - $H_0$: $z = \frac{\hat{y} - \mu_0}{\sqrt{\frac{\sigma_1^2}{N_1} + \frac{\sigma_2^2}{N_2}}} \sim N(0, 1)$

- Assume $\sigma_1 = \hat{\sigma_1}$, $\sigma_2 = \hat{\sigma_2}$. **Note**: $\hat{y} = \hat{\mu_1} - \hat{\mu_2}$
    - $H_0$: $z = \frac{\hat{y} - \mu_0}{\sqrt{\frac{\hat{\sigma_1}^2}{N_1} + \frac{\hat{\sigma_2}^2}{N_2}}} \sim N(0, 1)$
    
- Calculate p-value
    - $p = 1 - \text{norm.cdf}(z)$
