# 1. Continuous random variables
- Use cases of Discrete vs Continous Random variables

<img src="assets/9.png" width="600"/>

#### Continous random variables
- We interested in a segment value of X rather than a specific value
    + Probability of a segment (from a to b) value of X: $0 < P(X \in [a,b]) \leq 1$
    + Probability of $X = x_0$: $P(X = x_0) = 0$

<img src="assets/10.png" width="600"/>

## 1.1 Example
- Uniform distribution on \[0,1\]:
    + $P(X \in [a,b]) = b - a$, $0 \leq a < b \leq 1$
        + $P(X \in [0,1]) = 1$
    + $P(X < 0 \cup X > 1) = 0$

# 2. Probability density function (PDF)

## 2.1 Definition

<img src="assets/11.png" width="600"/>

- $pdf_X (x_0) = \lim\limits_{\Delta X \to 0} \frac{P(X \in [x_0, x_0 + \Delta x])}{\Delta x}$

## 2.2 Example
- Find pdf of Uniform distribution on \[0,1\]

#### Solve
- $\forall x_0 \in (0,1)$

$$pdf_X (x_0) = \lim\limits_{\Delta X \to 0} \frac{P(X \in [x_0, x_0 + \Delta x])}{\Delta x} = \frac{x_0 + \Delta x - x_0}{\Delta x} = \frac{\Delta x}{\Delta x} = 1$$

- $\forall x_0, x_0 < 0\ or\ x_0 > 1$

$$pdf_X (x_0) = \lim\limits_{\Delta X \to 0} \frac{P(X \in [x_0, x_0 + \Delta x])}{\Delta x} = \frac{0}{\Delta x} = 0$$

- $x_0 = 0$ or $x_0 = 1$, Discontinous => pmf not exist

## 2.3 Histogram
- Assume we have a random variable X, and data samples from X as $[x_1, x_2, x_3, \dots, x_n]$
- We can approximate $pdf_X$ as the histogram of X
    + As N large, and $| X_{k+1} - X_k | $ small, the histogram are more accurate to present $pdf_X$

# 3. Cumulative distribution function (CDF)
## 3.1 Definition

<img src="assets/13.png" width="600"/>

- $cdf_X(x) = P(X \leq x)$


## 3.2 Example

- Find cdf of Uniform distribution on \[0,1\]

#### Solve

$$cdf_X(x) = \begin{cases}
0\text{, if x < 0} \\
x\text{, if } x \in [0,1] \\
1\text{, if } x > 1
\end{cases}$$

## 3.3 CDF for Discrete Random Variable

$$cdf_X(x) = \sum\limits_{i=1}^N p(X = x_i)$$


#### Example
- Discrete Random Variable X has following pmf

| **X**      | **X=1** | **X=2** | **X=2.5** |
|------------|---------|---------|-----------|
| **P(X=x)** | 0.2     | 0.5     | 0.3       |

- Find $cdf_X$

#### Solve

$$cdf_X(x) = \begin{cases}
0\text{, if x < 1} \\
P(X=1) \text{, if } x \in [1,2) \\
P(X=1) + P(X=2)\text{, if } x \in [2,2.5) \\
P(X=1) + P(X=2) + P(X=2.5) \text{, if } x \in [2.5, \infty) \\
\end{cases}$$

$$=> cdf_X(x) = \begin{cases}
0\text{, if x < 1} \\
0.2 \text{, if } x \in [1,2) \\
0.7\text{, if } x \in [2,2.5) \\
1 \text{, if } x \in [2.5, \infty) \\
\end{cases}$$

<img src="assets/15.png" width="600"/>

## 3.4 Properties
- 1) cdf is non-strictly increasing
    + if $x_1 \leq x_2$, then $cdf_X(x_1) \leq cdf_X(x_2)$
- 2) $\lim\limits_{x \to -\infty} cdf_X(x) = 0$, $\lim\limits_{x \to +\infty} cdf_X(x) = 1$
- 3) $cdf_X(x) \in [0,1]$


# 4. Linking PDF and CDF
- Let X is the random variable. Define
    + $pdf_X(x) = p(x)$
    + $cdf_X(x) = F(x)$
    
## 4.1 cdf, pdf, and P(X)
- Probability of segment $X \in [a,b]$:
    + $P(X \in [a,b] ) = F(b) - F(a)$
- pdf = derivative of cdf
    + $p(x) = \lim\limits_{\Delta x \to 0} \frac{F(x + \Delta x) - F(x)}{\Delta x} = \lim\limits_{\Delta x \to 0} \frac{\Delta F}{\Delta x} = F'(x)$
- cdf = antiderivative of pdf
    + $F(b) - F(a) = \int\limits_a^b F'(x)dx = \int\limits_a^b p(x)dx = P(X \in [a,b] )$
    + $F(x) = P(X \leq x) = \int\limits_{-\infty}^x p(x)dx $

<img src="assets/16.png" width="600"/>

#### Properties
- $\int\limits_{-\infty}^\infty p(x)dx = 1$


## 4.2 cdf and percentile

- Example
    - A population has normal random distribution of height X ~ norm($\mu, \sigma$) with
        + mean = 170 cm
        + std = 7cm 

#### cdf
- cdf(x) = the probability(Or percentile) that X less or equal to x

$$F(x) = P(X \leq x) = \int\limits_{-\infty}^x p(x)dx $$

- Example: you are 160 cm tall, what percentile are you in?
    + `Ans`: $F(x=160)$

- Example: you are 160 cm tall, what is the probability that someone is taller than you?
    + `Ans`: $1 - F(x=160)$

#### Percentile (ppf)
- Percentile = percent point function (ppf)
    + = the value x at probability (or percentile) p
    + = Inverse function of cdf

$$ ppf(p) = F^{-1}(p)$$

- Example: at what height are you in the 95th percentile?
    + `Ans`: $ppf(p=0.95) = F^-1(p=0.95)$
    

## 4.3 Example: Uniform distribution on `[a,b]`
- Define $p(x) = pdf_X(x)$
+ We have $p(x) = \int\limits_{-\infty}^\infty p(x)dx = 1 = \int\limits_a^b p_0dx$ => $p_0 = \frac{1}{b-a}$
    
    
$$p(x) = \begin{cases}
    0,\ x \notin [a,b] \\
    \frac{1}{b-a},\ x \in (a,b)

# 6. Expected value of continuous random variable
## 6.1 Definition
- Let X is the continuous  random variable: $X$ ~ $pdf_X(x)=p(x)$
- E(X) is defined as

$$E(X) = \int\limits_{-\infty}^{\infty}xp(x)dx$$

## 6.2 Example
- Find E(X) of X ~ Uniform(0,1)
- Given

$$p(x) = \begin{cases}
    0,\ x < 0 \\
    1,\ x \in [0,1] \\
    0,\ x > 1
\end{cases}$$

#### Solve

$$\begin{split}
E(X) &= \int\limits_{-\infty}^{\infty}xp(x)dx = 0 + \int\limits_0^1 1*xdx + 0 \\
    &= [\frac{1}{2} x^2]_0^1 = \frac{1}{2}
\end{split}$$

## 6.3 Properties

- If the pdf is symmetric at $x_0$, then $E(X) = x_0$

# 7. Variance of continuous random variable
## 7.1 Formulas

- Let X is the continuous  random variable: $X$ ~ $pdf_X(x)=p(x)$

- continuous random variable variance has the same properties as discrete random variable

- $\text{Var}(X) = E[ (X - E(X))^2 ]) = E(X^2) -E(X)^2$


#### Expected value
- $E(Z) = E[f(X)] = \int\limits_{-\infty}^{\infty} f(x)p(x)dx$

- Let $c, c_1, c_2$ = constant; X,Y,Z = random variables
    - $E(Z = c) = c$
    - $E(Z = X + c) = E(X) + c$
    - $E(Z = cX) = cE(X)$
    - $E(Z = c_1X + c_2) = c_1E(X) + c_2$
    - $E(Z = X + Y) = E(X) + E(Y)$
    - $E(Z = X*Y) = \int\limits_{x}\int\limits_{y}x*y*\text{pdf}_{X,Y}(x,y) dx dy$
    - $E(Z = X*Y) = E(X)E(Y)$, Only If `X and Y are independent`
    - $E[Z = E(X)] = E(X)$
    - $E[Z = XE(Y)] = E(X)E(Y)$
    
#### Variance
- Let $c, c_1, c_2$ = constant; X,Y,Z = random variables
    - $Var(Z = c) = 0$
    - $Var(Z = X + c) = Var(X)$
    - $Var(Z=cX) = c^2 Var(X)$
    - $Var(Z=c_1X + c_2) = c_1^2 Var(X)$
    - $Var(Z = X + Y) = Var(X) + Var(Y) + 2 cov(X,Y)$
    - $Var(Z = X + Y) = Var(X) + Var(Y)$, Only If `X and Y are independent`

## 7.2 Intuition of Variance
- If Var(Y) > Var(X) then Y will spread wider than X

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

X_mean = 0
X_variance = 1
X_std = math.sqrt(X_variance)
x = np.linspace(X_mean - 3*X_std, X_mean + 3*X_std, 100)
plt.plot(x, stats.norm.pdf(x, X_mean, X_std), label='pdf_X')


Y_mean = 1
Y_variance = 5
Y_std = math.sqrt(Y_variance)
y = np.linspace(Y_mean - 3*Y_std, Y_mean + 3*Y_std, 100)
plt.plot(y, stats.norm.pdf(y, Y_mean, Y_std), label='pdf_Y')

plt.legend()
plt.show()
```

# 8. Joint CDF and PDF
- Let X and Y are 2 random variable defined on common probability space

## 8.1 Joint CDF
- Joint cdf of X and Y are defined as

$$cdf_{X,Y}(x,y) = P(X \leq x \cap Y \leq y)$$

## 8.2 Joint PDF
- Joint pdf of X and Y are defined as

$$pdf_{X,Y}(x_0, y_0) = \lim\limits_{\Delta x \to 0 \\ \Delta y \to 0} \frac{P(X \in [x_0, x_0 + \Delta x] \cap Y \in [y_0, y_0 + \Delta y] )}{\Delta x \Delta y} $$

#### Marginal probability

- $pdf_X(x) = \int\limits_{-\infty}^{\infty} pdf_{X,Y}(x,y)dy$
- $pdf_Y(y) = \int\limits_{-\infty}^{\infty} pdf_{X,Y}(x,y)dx$

## 8.3 Conditional PDF
- p = pdf

#### Joint Distribution

$$p(x,y)$$

#### Marginal distribution

$$p(x) = \int\limits_{-\infty}^{\infty} p(x,y)dy = \int\limits_{-\infty}^{\infty} p(y)p(x|y)dy$$
$$p(y) = \int\limits_{-\infty}^{\infty} p(x,y)dx = \int\limits_{-\infty}^{\infty} p(x)p(y|x)dx$$

#### Conditional distribution

$$p(x|y) = \frac{p(x,y)}{p(y)} = \frac{p(x)p(y|x)}{\int\limits_{-\infty}^{\infty} p(x,y)dx} = \frac{p(x)p(y|x)}{\int\limits_{-\infty}^{\infty} p(x)p(y|x)dx}$$
$$p(y|x) = \frac{p(x,y)}{p(x)} = \frac{p(y)p(x|y)}{\int\limits_{-\infty}^{\infty} p(x,y)dy} = \frac{p(y)p(x|y)}{\int\limits_{-\infty}^{\infty} p(y)p(x|y)dy}$$

# 9. Independence, covariance and correlation of continuous random variables

- Let X and Y are 2 random variable defined on common probability space

## 9.1 Independence
- X and Y are independent if 1 of the following conditions meet
    + $cdf_{X,Y}(x,y) = cdf_X(x)cdf_Y(y)$
    + $pdf_{X,Y}(x,y) = pdf_X(x)pdf_Y(y)$

## 9.2 covariance and correlation

$$cov(X, Y) = E [ (X- E(X)) (Y - E(Y))    ] = E(XY) - E(X)E(Y)$$

$$corr(X,Y) = \frac{cov(X,Y)}{\sqrt{Var(X)Var(Y)}}$$

#### Properties
- if `X and Y are independent` then $corr(X,Y) = cov(X, Y) = 0$, **Note**: The inverse is not true
