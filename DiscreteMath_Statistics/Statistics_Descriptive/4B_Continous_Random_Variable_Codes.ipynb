# 0. Histogram
- Count the frequency of data

```python
data = [1,1.2,4.1,4.2,4.3,5]

# Plot hist in 3 bins
#     1st bin [1,2.33): [1, 1.2] -> count = 2
#     2nd bin [2.33,3.66): [] -> count = 0
#     3rd bin [3.66,5.0]: [4.1,4.2,4.3,5] -> count = 3
plt.hist(x=data, bins=3)
plt.plot(data, len(data)*[1], 'o' )
plt.show()
```

#### Histogram with density instead of frequency

```python
plt.hist(x=data, bins=3)
plt.plot(data, len(data)*[1], 'o')
plt.show()
```

# 1. Uniform Distribution` [a,b]`

#### Simulation - Histogram

```python
from numpy.random import uniform

a = -2
b = 2

data = uniform(low=a, high=b, size=10)
print(data)
```

```python
data = uniform(low=a, high=b, size=100000)
plt.hist(x=data, bins=10, density=True)
plt.show()
```

#### By scipy.stats.uniform - Draw pdf

```python
from scipy.stats import uniform

a = -2
b = 2
len_ = b-a

X = uniform(a, len_)

# pdf_X(1.2)
print(X.pdf(1.2))
```

```python
x = np.linspace(-3,3,1000)
y = X.pdf(x)

plt.plot(x,y)
plt.show()
```

# 2. Generate correlated random variables

```python
from numpy.random import multivariate_normal
```

```python
def generate_correlated_data(
        X_mean, X_var,
        Y_mean, Y_var,
        cov_XY, datapoints=1000):

    cov_matrix = [
        [X_var, cov_XY],
        [cov_XY, Y_var]]

    data = multivariate_normal([X_mean, Y_mean], cov_matrix, size=datapoints)

    X = data[:,0]
    Y = data[:,1]

    plt.scatter(x=X, y=Y)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
```

```python
# Generate X~N(0,1), Y~N(0,1) with cov(X,Y) = 0.5
generate_correlated_data(
    X_mean=0, X_var=1,
    Y_mean=0, Y_var=1, cov_XY=0.5)
```

```python
# Generate X~N(0,1), Y~N(0,1) with cov(X,Y) = 1
generate_correlated_data(
    X_mean=0, X_var=1,
    Y_mean=0, Y_var=1, cov_XY=1)
```

```python
# Generate X~N(4,1), Y~N(0,1) with cov(X,Y) = -0.7
generate_correlated_data(
    X_mean=4, X_var=1,
    Y_mean=0, Y_var=1, cov_XY=-0.7)
```
